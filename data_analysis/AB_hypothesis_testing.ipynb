<h1 style="color:DarkSlateBlue" >6. A/B Testing</h1>


An A/B test is a method of comparing two versions of something to determine which performs better. In a game context, it involves testing two different game mechanics, features, or experiences (Version A and Version B) with different groups of players. The goal is to measure how changes affect player behavior, such as their ability to reach specific in-game gates or milestones.




# A/B Testing Workflow

## 1. Hypotheses
- **H₀ (Null Hypothesis)**: The target variable (e.g., `sum_gamerounds`) is the same for Groups A and B.
- **H₁ (Alternative Hypothesis)**: The target variable is different for Groups A and B.

---

## 2. Steps in A/B Testing

### Step 1: Data Splitting
- **What we do**: Split the data into two groups based on the `group` column: `groupA` and `groupB`.
- **Why we do this**: To analyze the target variable for each group separately.

---

### Step 2: Check Assumptions

#### a. Test for Normality (Shapiro-Wilk Test)
- **What it does**: Checks if the target variable in each group follows a normal distribution.
- **Why**: Determines whether we can use parametric tests (like t-test) or need non-parametric tests (like Mann-Whitney U).
- **Decision**:
  - If **p-value ≥ 0.05** → Data is normal → Use parametric tests.
  - If **p-value < 0.05** → Data is not normal → Use non-parametric tests.

#### b. Test for Homogeneity of Variances (Levene’s Test)
- **What it does**: Checks if the variances of the two groups are equal.
- **Why**: Determines whether to use a standard t-test (equal variances) or Welch’s t-test (unequal variances).
- **Decision**:
  - If **p-value ≥ 0.05** → Variances are equal → Use standard t-test.
  - If **p-value < 0.05** → Variances are not equal → Use Welch’s t-test.

---

### Step 3: Perform Statistical Tests

#### Case 1: Data is Normal in Both Groups
- **Check Homogeneity (Levene’s Test)**:
  - If variances are equal → Perform **standard t-test**.
  - If variances are unequal → Perform **Welch’s t-test**.
- **Hypotheses**:
  - H₀: Mean of Group A = Mean of Group B.
  - H₁: Mean of Group A ≠ Mean of Group B.

#### Case 2: Data is Not Normal in Any Group
- **Perform Mann-Whitney U Test**:
  - H₀: Median of Group A = Median of Group B.
  - H₁: Median of Group A ≠ Median of Group B.

---

### Step 4: Output Results

#### Results Summary:
- **Test Type**: Whether the test was parametric or non-parametric.
- **Homogeneity**: If variances are equal (if applicable).
- **A/B Hypothesis**: Whether H₀ was rejected or not.
- **p-value**: Indicates the strength of evidence against H₀.
- **Comment**: Explains whether the groups are similar or different.

---

## Key Takeaways
1. **Assumptions Matter**: Test choice depends on:
   - Normality (Shapiro-Wilk test).
   - Homogeneity of variances (Levene’s test).

2. **Hypothesis Testing Outcomes**:
   - **Reject H₀** → Groups are **different**.
   - **Fail to reject H₀** → Groups are **similar**.

3. **Decision Flow**:
   - Normal data → Parametric tests (t-test).
   - Non-normal data → Non-parametric tests (Mann-Whitney U).



### Assumptions:
- Check normality
- If Normal Distribution, check homogeneity

### Steps:
- Split & Define Control Group & Test Group
- Apply Shapiro Test for normality
- If parametric apply Levene Test for homogeneity of variances
- If Parametric + homogeneity of variances apply T-Test
- If Parametric - homogeneity of variances apply Welch Test
- If Non-parametric apply Mann Whitney U Test directly

```python
# Define A/B groups
ab["version"] = np.where(ab.version == "gate_30", "A", "B")
ab.head()
```

```python
# A/B Testing Function - Quick Solution
def AB_Test(dataframe, group, target):
    
    # Packages
    from scipy.stats import shapiro
    import scipy.stats as stats
    
    # Split A/B
    groupA = dataframe[dataframe[group] == "A"][target]
    groupB = dataframe[dataframe[group] == "B"][target]
    
    # Assumption: Normality
    ntA = shapiro(groupA)[1] < 0.05
    ntB = shapiro(groupB)[1] < 0.05
    # H0: Distribution is Normal! - False
    # H1: Distribution is not Normal! - True
    
    if (ntA == False) & (ntB == False): # "H0: Normal Distribution"
        # Parametric Test
        # Assumption: Homogeneity of variances
        leveneTest = stats.levene(groupA, groupB)[1] < 0.05
        # H0: Homogeneity: False
        # H1: Heterogeneous: True
        
        if leveneTest == False:
            # Homogeneity
            ttest = stats.ttest_ind(groupA, groupB, equal_var=True)[1]
            # H0: M1 == M2 - False
            # H1: M1 != M2 - True
        else:
            # Heterogeneous
            ttest = stats.ttest_ind(groupA, groupB, equal_var=False)[1]
            # H0: M1 == M2 - False
            # H1: M1 != M2 - True
    else:
        # Non-Parametric Test
        ttest = stats.mannwhitneyu(groupA, groupB)[1] 
        # H0: M1 == M2 - False
        # H1: M1 != M2 - True
        
    # Result
    temp = pd.DataFrame({
        "AB Hypothesis":[ttest < 0.05], 
        "p-value":[ttest]
    })
    temp["Test Type"] = np.where((ntA == False) & (ntB == False), "Parametric", "Non-Parametric")
    temp["AB Hypothesis"] = np.where(temp["AB Hypothesis"] == False, "Fail to Reject H0", "Reject H0")
    temp["Comment"] = np.where(temp["AB Hypothesis"] == "Fail to Reject H0", "A/B groups are similar!", "A/B groups are not similar!")
    
    # Columns
    if (ntA == False) & (ntB == False):
        temp["Homogeneity"] = np.where(leveneTest == False, "Yes", "No")
        temp = temp[["Test Type", "Homogeneity","AB Hypothesis", "p-value", "Comment"]]
    else:
        temp = temp[["Test Type","AB Hypothesis", "p-value", "Comment"]]
    
    # Print Hypothesis
    print("# A/B Testing Hypothesis")
    print("H0: A == B")
    print("H1: A != B", "\n")
    
    return temp
    

# Apply A/B Testing
AB_Test(dataframe=ab, group = "version", target = "sum_gamerounds")
```
