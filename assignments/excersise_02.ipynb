# تمرینات کار با فایل `housing.csv`

## 1. خواندن داده‌ها
فایل `housing.csv` را با استفاده از pandas بخوانید و داده‌ها را در یک DataFrame ذخیره کنید.

## 2. نمایش اطلاعات کلی
اطلاعات کلی داده‌ها مانند تعداد سطرها، ستون‌ها، و نوع داده‌ها را با استفاده از متد `.info()` بررسی کنید.

## 3. توزیع مقادیر
توزیع داده‌ها در ستون‌های عددی را با استفاده از متد `.describe()` بررسی کنید.

## 4. بررسی داده‌های گمشده
ستون‌هایی که داده‌های گمشده دارند را شناسایی کنید و تعداد آن‌ها را در هر ستون نشان دهید.

## 5. فیلتر کردن داده‌ها
فیلتر کنید تا فقط خانه‌هایی که قیمت آن‌ها بالاتر از 500 هزار دلار است نمایش داده شوند.

## 6. افزودن ستون جدید
یک ستون جدید به نام `price_per_room` اضافه کنید که نشان‌دهنده نسبت قیمت به تعداد اتاق‌ها باشد.

## 7. مرتب‌سازی داده‌ها
داده‌ها را بر اساس ستون قیمت مرتب کنید و 10 خانه با بالاترین قیمت را نمایش دهید.

## 8. رسم هیستوگرام
هیستوگرام قیمت خانه‌ها را رسم کنید تا توزیع قیمت‌ها را بررسی کنید.

## 9. نمودار پراکندگی
یک نمودار پراکندگی (scatter plot) رسم کنید که قیمت خانه‌ها را بر اساس مساحت زمین نمایش دهد.

## 10. رسم نقشه جغرافیایی
موقعیت خانه‌ها را بر اساس مختصات جغرافیایی (latitude و longitude) روی یک نمودار نقطه‌ای نمایش دهید.
