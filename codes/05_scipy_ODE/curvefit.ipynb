# Curve Fit

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit

# Define the functions
def quadratic_function(x, a, b, c):
    return a * x**2 + b * x + c

def exponential_function(x, a, b):
    return a * np.exp(b * x)

def sine_function(x, a, b):
    return a * np.sin(b * x)

def logarithmic_function(x, a, b):
    return a * np.log(b * x)

def cubic_function(x, a, b, c, d):
    return a * x**3 + b * x**2 + c * x + d

# Generate x values
x = np.linspace(0, 10, 100)

# Generate data points with noise
y_quadratic = quadratic_function(x, 2, 3, 5) + np.random.normal(0, 2, len(x))
y_exponential = exponential_function(x, 2, 0.5) + np.random.normal(0, 1, len(x))
y_sine = sine_function(x, 3, 0.8) + np.random.normal(0, 1, len(x))
y_cubic = cubic_function(x, 1, -2, 3, 4) + np.random.normal(0, 3, len(x))

# Perform curve fitting
params_quadratic, _ = curve_fit(quadratic_function, x, y_quadratic)
params_exponential, _ = curve_fit(exponential_function, x, y_exponential)
params_sine, _ = curve_fit(sine_function, x, y_sine)
params_cubic, _ = curve_fit(cubic_function, x, y_cubic)

# Plot the results
plt.figure(figsize=(12, 8))

plt.subplot(2, 3, 1)
plt.scatter(x, y_quadratic, label='Data with Noise')
plt.plot(x, quadratic_function(x, *params_quadratic), 'r', label='Fitted Curve')
plt.title('Quadratic Function')
plt.legend()

plt.subplot(2, 3, 2)
plt.scatter(x, y_exponential, label='Data with Noise')
plt.plot(x, exponential_function(x, *params_exponential), 'r', label='Fitted Curve')
plt.title('Exponential Function')
plt.legend()

plt.subplot(2, 3, 3)
plt.scatter(x, y_sine, label='Data with Noise')
plt.plot(x, sine_function(x, *params_sine), 'r', label='Fitted Curve')
plt.title('Sine Function')
plt.legend()

plt.subplot(2, 3, 5)
plt.scatter(x, y_cubic, label='Data with Noise')
plt.plot(x, cubic_function(x, *params_cubic), 'r', label='Fitted Curve')
plt.title('Cubic Function')
plt.legend()

plt.tight_layout()
plt.show()
```
